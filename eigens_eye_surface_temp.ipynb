```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, scipy.optimize
from scipy.optimize import root_scalar
```

```python
# define equation
def eqns(x):

    # Define constants
    B = 0.3

    # Define function
    f = np.tan(x) + x / B
    return f

def eqnsPrime(x):
    B = 0.3
    return (1 / np.cos(x)) ** 2 + 1 / B
```

```python
# Set initial guesses. This is different from MATLAB that 2 guesses are required for secant method.
x0 = 9
x1 = 8
#call root_scalar(eqns)
result = root_scalar(eqns,x0 = x0, fprime = eqnsPrime, method = 'newton')
print("newtown method:", result.root)
result = root_scalar(eqns,x0 = x0, x1 = x1, method = 'secant')
print("secant method:", result.root)
result = root_scalar(eqns,bracket = [-8 , 8], method = 'bisect')
print("bisect method:", result.root)
result = scipy.optimize.root(eqns, 8, method = 'lm')
print("lm:", result.x)
print("The result is different depend on the method and function you choose. \n The most method closest to MATLAB is using scipy.optimize.root and using method lm")
```

```python
# Plot of function
x = np.arange(-15,15,0.01)
y = np.tan(x) + x / 0.3
plt.figure()
plt.plot(x,y)
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.xlabel("x")
plt.ylabel("y")
plt.title('plot of y = tan(x) + x / 0.3')
plt.show()
```
